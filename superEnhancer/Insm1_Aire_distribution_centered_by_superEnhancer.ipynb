```julia
include("../deps/dataProcessKit/dataProcessKit.jl")
using .dataProcessKit
include("../deps/dataProcessKit/myDraw.jl")
using .myDraw
```

```julia
se=jload("super_enhancer_mTECs_providedByAuthor")
se["scan_pos"]=round.(Int, meanh(se["pos"]))|>x->[x.-800000 x.+800000]
pkI=jload("../Insm1_cut_and_tag/merged_Insm1_peak_E18.jld2")
pkA=jload("../Insm1_cut_and_tag/Aire_ChIP_WT_4rep_narrowPeak.jld2")
Ii, Ai=genomemap(d"chr, pos"pkI, d"chr, pos"pkA, touch=true)
pkI["isAire"]=falsesbut(rnum(pkI), Ii)c
pkIu=rec(pkI, .!pkI["isAire"])
pkIc=rec(pkI, pkI["isAire"])
pkAu=rec_i(pkA, Ai)

si, pi=genomemap(d"chr, scan_pos"se, (pkIu["chr"], pkIu["abs_summit"]|>x->[x x]))
distIu=round.(Int, meanh(se["pos"]))[si].-pkIu["abs_summit"][pi]
rIu=uninum(pi)/rnum(pkIu)

si, pi=genomemap(d"chr, scan_pos"se, (pkIc["chr"], pkIc["abs_summit"]|>x->[x x]))
distIc=round.(Int, meanh(se["pos"]))[si].-pkIc["abs_summit"][pi]
rIc=uninum(pi)/rnum(pkIc)

si, pi=genomemap(d"chr, scan_pos"se, d"chr, summit_pos"pkAu)
distAu=round.(Int, meanh(se["pos"]))[si].-pkAu["summit_pos"][pi, 1];
rAu=uninum(pi)/rnum(pkAu);
```

```julia
figure(figsize=(5, 4))
pdensity((distIc, distIu, distAu), times=[rIc, rIu, rAu], legend=c"Insm1/Aire-cobound, Insm1-only, Aire-only")
ylim(0, ylim()[2])
xlim(-600000, 600000)
guideline("|", "-", color="0.8")
ylabel("Peak density")
xlabel("Distance to super enhancer center (bp)")
xticks()[1]|>x->xticks(x, f"$1k".(round.(Int, x./1000)))
yticks([], [])
ticklabelreplace(:x, "0k"=>"0")
```
