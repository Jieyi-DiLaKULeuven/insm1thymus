```julia
#K=let cAD=ADa, order=orderls_a
K=let cAD=ADe, order=orderls_e
    T=pandas2tb(cAD.obs)
    ge="Aire" #"Insm1"
    #ge="Insm1"
    T["geexp"]=sc.get.obs_df(cAD, ge, use_raw=true).to_list()
    K=grpfun(fastgrp(T["celltype"], order), 1:rnum(T)) do l
        nl=truesbut(rnum(T), l)
        A=T["geexp"][l]
        B=T["geexp"][nl]
        P2=wilcoxtest(A, B)
        P1=wilcoxtest(A, B, alternative="greater")
        (a, b, c, d)=(count(A.>0), count(B.>0), count(A.==0), count(B.==0))
        oddsratio=(a*d)/(b*c)
        Pf=fishertest([a b; c d])
        (mean(A), mean(B), median(A), median(B), P2, P1, mean(A.>0), mean(B.>0), oddsratio, Pf)
    end |> tb(c"mean, meanCT, median, medianCT, Pvalue_2tails, Pvalue_1tails, expressedIn%cell, expressedIn%CTcell, OddsRatio, FisherPvalue");
    K["celltype"]=order
    @with K $meanDiff=$mean.-$meanCT
    @with K $medianDiff=$median.-$medianCT;
    K
end
tocb(asmx(c"celltype, mean, meanCT, median, medianCT, Pvalue_2tails, Pvalue_1tails, expressedIn%cell, expressedIn%CTcell, OddsRatio, FisherPvalue"=>K))
```
