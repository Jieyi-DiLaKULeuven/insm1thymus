```julia
include("../deps/dataProcessKit/dataProcessKit.jl")
using .dataProcessKit
include("../deps/dataProcessKit/myDraw.jl")
using .myDraw
```

```julia
se=jload("super_enhancer_mTECs_providedByAuthor")
pk=jload("../Insm1_cut_and_tag/merged_Insm1_peak_E18.jld2")

si, pi=genomemap(d"chr, pos"se, d"chr, pos"pk, touch=true)
pk["isSupEnh"]=falsesbut(rnum(pk), pi)
se["isInsm1"]=falsesbut(rnum(se), si)

ge=jload("../Insm1_KO_RNAseq/E18_gene_DESeq2_test.jld2")
ge=rec(ge, (ge["raw_pvalue"].<0.05) .& (abs.(ge["logFC"]).>log2(1.5)))
#ge=rec(ge, ge["isDE"])
se["scan_pos"]=se["pos"]|>x->[x[:, 1].-500000 x[:, 2].+500000]
si, gi=genomemap(d"chrno, scan_pos"se, d"chrno, gene_pos"ge, touch=true)
se["isDEge"]=falsesbut(rnum(se), si);
```

```julia
#Add control
se["ct_pos"]=se["pos"].+seglen(se["pos"]).+200000
si, pi=genomemap(d"chr, ct_pos"se, d"chr, pos"pk, touch=true)
pk["ct_isSupEnh"]=falsesbut(rnum(pk), pi)
se["ct_isInsm1"]=falsesbut(rnum(se), si);
```

```julia
figure(figsize=(2.5, 4))
@with pk begin
    _, _, tx, ty=binobar($isSupEnh, $ct_isSupEnh)
    halfbox()
    p=fishertest_l($isSupEnh, $ct_isSupEnh)
    barbrace(tx[1:2], max(ty...)+ax_height(0.03), f"p=$(1=.4g)"(p), balance="<")
    xticks(1:2, c"super\nenhancer, Cntl.")
    ylabel("Proportion of Insm1 binding")
end
pltsave("Prop_of_Insm1_binding_SE_barplot")
```

```julia
figure(figsize=(2.5, 4))
_, _, tx, ty=@with se binobar($isDEge[.!$isInsm1], $isDEge[$isInsm1])
halfbox()
p=@with se fishertest_l($isDEge[.!$isInsm1], $isDEge[$isInsm1])
barbrace(tx, maximum(ty)+ax_height(0.03), f"p=$(1=.3g)"(p), balance=">")
xticks(1:2, c"Insm1-\nunbound, Insm1-\nbound")
ylabel("Super enhancer related genes")
pltsave("prop_SE_related_gene_barplot")
```
